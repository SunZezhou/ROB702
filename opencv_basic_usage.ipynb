# Basic Usage of OpenCV

```python
# Import Dependency Libraries

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Image Reading and Writing

# Read an Image
img = cv2.imread('resource/Lena.png', cv2.IMREAD_COLOR)   # Modes: IMREAD_COLOR, IMREAD_GRAYSCALE, IMREAD_UNCHANGED

# Convert to Grayscale or RGB Color Space
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Write an Image
cv2.imwrite('resource/output.png', img)

print(gray_img) # Width * Height * Channel
```

```python
# Display the image using matplotlib

# cv2.imshow('Image Display with OpenCV', img)
# plt.imshow(gray_img, cmap='gray')  
plt.imshow(color_img)
plt.axis('on')  # Optional: Hide axes for a cleaner look
plt.title('Image Display with Matplotlib')  # Optional: Add a title
plt.show()
```

```python
# Resize an Image

resized_img = cv2.resize(color_img, (640, 480)) # (Width, Height)

plt.imshow(resized_img)
```

```python
# Rotate an Image

(h, w) = img.shape[:2]
center = (w // 2, h // 2)
matrix = cv2.getRotationMatrix2D(center, 30, 0.7) # center angel(degree), scale
rotated_img = cv2.warpAffine(color_img, matrix, (w, h))

plt.imshow(rotated_img)
```

```python
# Draw Shapes on an Image

draw_img = color_img.copy()
# cv2.line(color_img, (150, 50), (200, 50), [100, 100, 255], 1) # cv2.line(img, (x1, y1), (x2, y2), color, thickness)
cv2.rectangle(draw_img, (150, 50), (200, 100), [100, 100, 255], 1) # cv2.rectangle(img, (x1, y1), (x2, y2), color, thickness)

plt.imshow(draw_img)
# plt.imshow(color_img)
```

```python
# Corp an Image

cropped_img = color_img[50:100, 150:200] # img[y1:y2, x1:x2]


plt.imshow(cropped_img)
```

```python
# Image Filtering

blurred_img = cv2.GaussianBlur(color_img, (5, 5), 0) # cv2.GaussianBlur(img, (kernel_width, kernel_height), sigma)

plt.imshow(blurred_img)
```

```python
# Image Thresholding

_, binary_img = cv2.threshold(gray_img, 100, 255, cv2.THRESH_BINARY)

plt.imshow(binary_img, cmap='gray')
```

```python
# Image Erosion and Dilation

erosion_img = cv2.erode(binary_img, np.ones((3, 3), np.uint8), iterations=1) # cv2.erode(img, kernel, iterations)

dilation_img = cv2.dilate(binary_img, np.ones((3, 3), np.uint8), iterations=1) # cv2.dilate(img, kernel, iterations)

kernel = np.ones((3, 3), np.uint8)
mask = cv2.morphologyEx(binary_img, cv2.MORPH_OPEN, kernel) # cv2.morphologyEx(img, operation, kernel)
mask = cv2.morphologyEx(binary_img, cv2.MORPH_CLOSE, kernel)

plt.imshow(mask, cmap='gray')
```

```python
# Find Contours

contours, _ = cv2.findContours(binary_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) # cv2.findContours(img, mode, method)
contours_img = color_img.copy()
cv2.drawContours(contours_img, contours, -1, (0, 255, 0), 1)

plt.imshow(contours_img)
```
